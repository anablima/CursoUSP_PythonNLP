## Funções NLTK

### Tokenizer

> Separação de palavras do texto e também expressões regulares.



```python
# Separar as palavras do texto
texto = 'A Ana tem 24 anos, é muito inteligente e gostosa!'
tokens = nltk.word_tokenize(texto)
print(tokens)
```

```python
# Separando as palavras do texto retirando as pontuações
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(texto)
print(tokens)
```

```python
# Separando as palavras do texto retirando os números
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[A-z]\w+')
tokens = tokenizer.tokenize(texto)
print(tokens)
```

### Frequência/Contagem

> Contagem de tokens.



```python
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(texto)
frequencia = nltk.FreqDist(tokens)
print(frequencia.most_common(5))
```

```python
# Tornando todas as palavras do texto em minusculo
new_list = []
for palavra in tokens:
  new_list.append(palavra.lower())

print(new_list)
```

```python
# Contando a frequência dessas palavras
frequencia = nltk.FreqDist(new_list)
frequencia.most_common()
```

```python
# Fazendo as duas últimas células com apenas 2 linhas :)
frequencia = nltk.FreqDist([palavra.lower() for palavra in tokens])
frequencia.most_common()
```

### StopWords

> São palavras que podem ser consideradas como irrelevantes para nosso projeto/análise.



```python
# Mostrando todas as palavras consideras irrelevantes nos textos do 'corpus'
nltk.corpus.stopwords.words('portuguese')
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')
texto = 'A Ana tem 24 anos, é muito inteligente e gostosa!'
tokens = tokenizer.tokenize(texto)
stopwords = nltk.corpus.stopwords.words('portuguese')
tokens_sem_stopwords = [w.lower() for w in tokens if w.lower() not in stopwords]
print('Palavras mais irrelevantes no texto:', tokens_sem_stopwords)

frequencia = nltk.FreqDist(tokens_sem_stopwords)
print('Frequência dessas palavras no texto:', frequencia.most_common())
```

> N-Gramas são palavras que aparecem em conjunto.

* Bigramas -> 2 palavras
* Trigramas -> 3 palavras
