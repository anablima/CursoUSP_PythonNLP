```python
import spacy

texto = "Na pandemia, por se tratar de uma situação excepcional, o período de graça pode ser prorrogado. Conforme previsto por lei, o trabalhador continua na condição de segurado ainda que não possa pagar as contribuições durante “até 12 (doze) meses após cessar a segregação, o segurado acometido de doença de segregação compulsória”. Entretanto, a legislação não define ao certo do que se trata essa segregação compulsória, nem se a necessidade de isolamento social devido à pandemia dá o direito a essa garantia. Desta forma, a situação pode ser uma justificativa para a proteção desse direito, mas não é uma garantia de que o segurado poderá passar um tempo sem contribuir com o INSS."

nlp = spacy.load('pt_core_news_lg')
doc = nlp(texto)
```

```python
tokens = [token for token in doc]
tokens
```

```python
tokens_orth = [token.orth_ for token in doc]
print([token.orth_ for token in doc if token.is_alpha])
print([token.orth_ for token in doc if token.is_digit])
print([token.orth_ for token in doc if token.is_punct])
print([token.orth_ for token in doc if not token.is_stop and token.is_alpha])
```

### Lematização

```python
print([token.lemma_ for token in doc if token.pos_ == "VERB"])
```

### Part-of-Speech Tagging

```python
etiquetado = [(token.orth_, token.pos_) for token in doc]
etiquetado
```

```python
morfologico = [(token.orth_, token.morph) for token in doc]
morfologico
```
